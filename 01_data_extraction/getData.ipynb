```python
import pyodbc
import pandas as pd
```

### Menghubungkan Python dengan SQL Server

#### Sesuaikan nama server, database, user, dan password dengan SQL Server yang dituju

```python
try:
    connection = pyodbc.connect('DRIVER={SQL SERVER};' +
                                'SERVER=NAMA_SERVER' +
                                'DATABASE=NAMA_DATABASE' +
                                'UID=USERNAME' +
                                'PWD=PASSWORD')
    print("Connection successful")

except pyodbc.Error as ex:
    print("Failed to connect")
```

### Mengambil Data Notice 2019

#### Query

##### Kolom-kolom ini akan digunakan untuk mengolah data notice, baik menghitung PKB ataupun BBN. Namun, apabila ingin mengambil data lain, maka bisa disesuaikan di bagian SELECT dengan memilih kolom apa saja yang diperlukan.
##### Query yang disesuaikan ada di bagian FROM untuk merujuk tabel mana yang akan digunakan dan juga WHERE di bagian YEAR(bnk_tanggal) yang dapat diganti sesuai tahun dari data yang akan diambil

```python
query = """
SELECT [row_id], [kd_drive], [knd_id],[ntc_tb_bbn1_p],[ntc_tb_bbn1_d],[ntc_tb_bbn2_p],[ntc_tb_bbn2_d], [ntc_tb_pkb_p], [ntc_tb_pkb_d],
       [ntc_tg_tahun], [ntc_tg_bulan], [ntc_tg_hari], [ntc_tg_pkb_p], [ntc_tg_pkb_d], [bnk_tanggal], [ctk_notice_tanggal], [knd_tgl_notice_old], 
       [knd_tgl_notice_new], [knd_tgl_stnk_start], [knd_tgl_stnk_end], [knd_nopol], [kab_desc], [kec_desc], [jns_desc], [mrk_desc], 
       [ntc_dasar_pkb], [kd_sex], [job_desc]
FROM trx_notice_19_22
WHERE YEAR(bnk_tanggal) = 2019 AND bnk_tanggal IS NOT NULL AND ntc_posisi = -5 AND ctk_notice = -1;
"""
```

### Menjalankan Query and Menyimpan Data dalam Bentuk CSV

#### Ganti nama file/path yang akan digunakan untuk menyimpan data di bagian "df.to_csv()"
#### Apabila ingin menyimpan file CSV di dalam folder yang sama dengan code ini berada, maka cukup tuliskan nama file saja, namun jika ingin menyimpan di folder yang berbeda, maka copy path tujuan yang disertai dengan nama file yang berekstensi .csv

```python
df = pd.read_sql(query, connection)
df.to_csv("nama_file_atau_path.csv", index=False)
print("Data berhasil disimpan dalam format CSV!")
```
